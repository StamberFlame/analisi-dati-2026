```python
import matplotlib.pyplot as plt
import numpy as np
import math

def logistic(x, r):
    return r*x*(1 - x)
    
def plot_logistic_sequence(x0, r, nn=101):
    x = x0
    px = []
    py = []
    for n in range(nn):
        px.append(n)
        py.append(x)
        x = logistic(x, r)
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.set_xlabel('$n$')
    ax.set_ylabel('$x_n$')
    ax.plot(px, py)
    print("x(",nn-1,")=",x)
    
plot_logistic_sequence(0.5,3.9,100)
```
